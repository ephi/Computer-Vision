```python
%matplotlib inline

import matplotlib.pyplot as plt
from scipy.misc import imread
import numpy as np
```

## Image thresholding

(A) Load the 'plate.png' image.

(B) Plot its values histogram.

(C) Threshold the images at 4 diffrent values (th=10, 20, 160 and 170), and display the resulting binary images (use subplots).

```python
img = imread('plate.png')
plt.imshow(img,cmap="gray",interpolation="nearest")
plt.colorbar(fraction=0.025)
plt.show()
```

```python
# Plotting Histogram
plt.hist(np.array(img).flatten(),bins=20,range=(0,255))
plt.show()

def thresholdImg(img,th):
    newImg = (img > th)
    newImg = newImg + 255
    return newImg
plt.figure(figsize=(13,7))
for i,th in enumerate([10,20,160,170]):
    plt.subplot(2,2,i+1)
    plt.title('img > {}'.format(th))
    plt.imshow(thresholdImg(img,th),cmap="gray",interpolation="nearest")
    plt.colorbar(fraction=0.025)
plt.show()
```

### rice.jpg

(A) Load and display 'rice.jpg' image and its histogram.

(B) Threshold the image to produce a good binary segmentation (between the coins and its background).

(C) Calculate and compare the histograms of the top-half of the image and the bottom-half of the image.

```python
img = imread('rice.jpg')
plt.imshow(img,cmap="gray",interpolation="nearest")
plt.colorbar(fraction=0.025)
plt.show()

```

```python
# Plotting Histogram
plt.figure(figsize=(9,8))
plt.subplot(2,2,1)
plt.title('Rice Full Histogram')
plt.hist(np.array(img).flatten(),bins=20,range=(0,255))

# Plotting Threshold Image
plt.subplot(2,2,2)
plt.title('Background Thresholding')
plt.imshow(thresholdImg(img,125),cmap="gray",interpolation="nearest")
plt.colorbar(fraction=0.025)

plt.subplot(2,2,3)
topHalfImg = img[0:127,:]
bottomHalfImg = img[128:255,:]
# Plotting Overlayed Histograms of Top Half and Bottom Half
plt.title('Top-Half Bottom-Half comparison')
plt.hist(np.array(topHalfImg).flatten(),bins=20,range=(0,255))
plt.hist(np.array(bottomHalfImg).flatten(),bins=20,range=(0,255))
plt.legend(['Top-Half','Bottom-Half'])
plt.show()
```
